```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
plt.title("Line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(a,b)
```

```python
day=np.arange(1,8)
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70,90,160,150,140,145,175]
plt.title("Line plot showing sale trends",color="green")
plt.xlabel("Days",color="green")
plt.ylabel("Sales",color="green")
plt.plot(day,sales_1,'ro--',label="company 1")
plt.plot(day,sales_2,'bo--',label="company 2")
plt.legend()
```
