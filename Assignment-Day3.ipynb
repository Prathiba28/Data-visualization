```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from numpy.random import randn
```

```python
sns.get_dataset_names()
```

```python
fmri=sns.load_dataset('fmri')
fmri.head(5)
```

```python
fmri.describe()
```

```python
fmri.shape
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',style='event',kind='line',data=fmri,aspect=3)
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',style='event',kind='scatter',data=fmri,aspect=3)
```

```python
plt.figure(figsize=[15,7])
df=sns.boxplot(x='timepoint',y='signal',hue='region',data=fmri,fliersize=3,dodge=True)
```

```python
my_colors = ["#9b59b6", "#3498db", 
             "#2ecc71", "#006a4e"]
sns.set_palette( my_colors )
plt.figure(figsize=[15,7])
df=sns.boxplot(x='timepoint',y='signal',hue='region',data=fmri,fliersize=3,dodge=False)
```
